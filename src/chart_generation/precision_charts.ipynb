```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.ticker import FuncFormatter

# Load the CSV file
file_path = '../results/perf/m_chart.csv'
file_path2 = '../results/perf/m_chart_2.csv'
data_df = pd.read_csv(file_path)
data_df_2 = pd.read_csv(file_path2)

# Extract x values from the first row (assuming the first row is the header)
x_values = data_df.columns.astype(float)
# Extract y values from the subsequent rows
y_values = data_df.values
y_values_2 = data_df_2.values


# Create a scatter plot
plt.figure(figsize=(10, 6))

for row in y_values:
    plt.scatter(x_values, row, color='blue')

for row in y_values_2:
    plt.scatter(x_values, row, color='red')

plt.legend(['t = 10000', 't = |E|'])
plt.xlabel('Rozmiar szkicu')
plt.ylabel('Precyzja')
plt.title('Precyzja w zależności od rozmiaru szkicu')
plt.xscale('log', base=2)
plt.grid(True)
plt.tight_layout()
ax = plt.gca()
leg = ax.get_legend()
leg.legendHandles[0].set_color('red')
leg.legendHandles[1].set_color('blue')

# Save the plot to a file and show it
plot_file_path = 'precision_m.png'
plt.savefig(plot_file_path)
plt.show()
```
